## Jump statement

### break - exit and  continue - skip

```python
for i in range(1,15):
    print(i)
    if i == 10:
        break
```

```python
n = 1

while(n<15):
    if n == 10:
        break
    print(n)
    n += 1
```

```python
for i in range(1,15):
     if i == 10:
         continue
     print(i)
```

```python
n = 1

while(n<15):
    if n == 10:
        n += 1
        continue
    print(n)
    n += 1
```

```python
for i in range(1,15):
     if i == 5:
         continue
     if i == 10:
         break
     print(i)
```

```python
n = 1

while(n<15):
    if n == 10:
        n += 1
        continue
    if n == 13:
        break
    print(n)
    n += 1
```

## Nested While loop

```python
n = 1

while n <= 5:  # 3 <=5 
    j = 1        # 2
    while j <= n:       # 1<= 1
        print(n,end=" ")             # 1
        j +=1              # 1+1 2
    print()
    n += 1                 # 1+1 2
```

```python
n = 1

while(n<=3):
    print(n)
    n += 1
```

```python
n = 1

while n <= 5:  # 3 <=5 
    j = 1        # 2
    while j <= n:       # 1<= 1
        print("*",end=" ")             # 1
        j +=1              # 1+1 2
    print()
    n += 1       
```

```python
num = 3

a = 1

while(a <= 10):  # 2
    print(num,"X",a,"=",num*a) # 3 x 2 = 3*2 =6
    a += 1   # 2
```
