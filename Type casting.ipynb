### Type casting is means changing the data type to another data type of same values

```python
num = 5            # int can be change a data type as str and float amd bool
print(num) 
print(type(num))
```

```python
num = str(5)   # priority
print(num)
print(type(num))
```

```python
num = "5"
print(num)
print(type(num))
```

```python
st = str(num)
print(st)
print(type(st))
```
