```python
groceries = ["apple", "banana", "orange", "pear"]

stock = {
    "banana": 6,
    "apple": 0,
    "orange": 32,
    "pear": 15
}

prices = {
    "banana": 4,
    "apple": 2,
    "orange": 1.5,
    "pear": 3
}

def computeBill(cart):
    total = 0

    for item in cart:
        if stock[item] > 0:
            total += prices[item]
            stock[item] -= 1
            print(f"{item} added to bill at ${prices[item]}")
        else:
            print(f"{item} is out of stock")

    return total
bill_amount = computeBill(groceries)
print("Total bill:", bill_amount)

```

```python
cart = {
    "apple": 2,
    "banana": 4,
    "orange": 3,
    "pear": 2
}
stock = {
    "banana": 6,
    "apple": 0,
    "orange": 32,
    "pear": 15
}

prices = {
    "banana": 4,
    "apple": 2,
    "orange": 1.5,
    "pear": 3
}
def computeBill(cart):
    total = 0

    for item, quantity in cart.items():

        if item not in prices:
            print(f"{item} not sold here")
            continue

        if stock[item] >= quantity:
            cost = prices[item] * quantity
            total += cost
            stock[item] -= quantity
            print(f"{item}: {quantity} × {prices[item]} = {cost}")

        else:
            print(f"Not enough stock for {item}")

    return total
bill = computeBill(cart)
print("Total bill:", bill)

```

```python
def number_stats(nums):
    tot = 0
    mini = nums[0]
    maxi = nums[0]

    for num in nums:
        tot = tot + num

        if num < mini:
            mini = num

        if num > maxi:
            maxi = num

    avg = tot / len(nums)

    return tot, avg, mini, maxi


numbers = [10, 5, 8, 3, 12]
tot, avg, mini, maxi = number_stats(numbers)

print("Total:", tot)
print("Average:", avg)
print("Minimum:", mini)
print("Maximum:", maxi)
```

```python
numbers = [10, 5, 8, 3, 12]

def number_stats(nums):
    total = sum(nums)
    average = total / len(nums)
    minimum = min(nums)
    maximum = max(nums)

    return total, average, minimum, maximum

result = number_stats(numbers)

print("Sum:", result[0])
print("Average:", result[1])
print("Minimum:", result[2])
print("Maximum:", result[3])
```

```python
number = [10, 5, 8, 3, 12]

def number_stats(nums):
    total = 0
    for num in nums:
        total += num

    sortvalue = sorted(nums)
    mini = sortvalue[0]
    maxi = sortvalue[-1]
    length = len(nums)
    averagevalue = total / length

    return total, mini, maxi, averagevalue

total, mini, maxi, average = number_stats(number)

print("Total:", total)
print("Minimum:", mini)
print("Maximum:", maxi)
print("Average:", average)
```

```python
# String & Set Analyzer

text = "hello world"

# Count vowels
vowels = "aeiou"
vowel_count = 0

for ch in text:
    if ch in vowels:
        vowel_count += 1

print("Number of vowels:", vowel_count)

# Create a set of unique characters (ignore spaces)
unique_chars = set()

for ch in text:
    if ch != " ":
        unique_chars.add(ch)

print("Unique characters:", unique_chars)
```
