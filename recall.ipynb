### Nested While loop 

```python
n = 1

while n <= 5:
    m = 1
    while m <= n:
        print(n,end = " ")
        m += 1
    print()
    n += 1
```

```python
n = 1

while n <= 5:
    m = 1
    while m <= n:
        print(m,end = " ")
        m += 1
    print()
    n += 1
```

```python
# n = 1
# n < = 5 -> 1<=5 -> T

# m = 1
# m<= n -> 1 <= 1 -> T
# m -> 1

#m += 1 -> 1+=1 =2 -> m = 2

# m = 2
# m<= n -> 2 <= 1 -> F

# n+=1 1+1 = 2 -> n = 2
```

```python
# n = 2
# n < = 5 -> 2<=5 -> T

# m = 1     # reset to IntiAL VALUE
# m<= n -> 1 <= 2 -> T
# m -> 1 -> P

#m += 1 -> 1+=1 =2 -> m = 2

# m = 2
# m<= n -> 2 <= 2 -> T
# m -> 2 -> P
#m += 1 -> 2+=1 -> 3

#m = 3
#m <=n -> 3<=2 -> F

# n+=1 2+1 = 3 -> n = 3
```

```python
n = 1

while n <= 5:
    m = 1
    while m <= n:
        print("*",end = " ")
        m += 1
    print()
    n += 1
```

```python
n = 5

while n >=1:
    m = 1
    while m<=n:
        print("*",end=" ")
        m+=1
    print()
    n -= 1
```

```python
n = 1

while n <= 5:
    m = 1
    while m<=5:
        print("*",end = " ")
        m += 1
    print()
    n += 1  
```

```python
n = 1

while n <=5: # 1 <=5 -> T
    gap = 5 - n   #5-1 -> 4
    while gap>0:   # 4>0 -> T
        print(" ",end=" ")
        gap-=1      # 4-1 -> 3
    pattern = 2*n-1  # 1
    while pattern>0:
        print("*",end=" ")
        pattern -=1
    print()
    n+=1
```
