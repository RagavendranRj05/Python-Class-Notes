# Average

### Method 1

```python
tamil_mark,engish_mark,maths_mark,science_mark,socail_mark = 90,70,80,99,98
print(tamil_mark,engish_mark,maths_mark,science_mark,socail_mark)
```

```python
marks = tamil_mark+engish_mark+maths_mark+science_mark+socail_mark
print(marks)
```

```python
avg_mark = marks/5
print(avg_mark)
```

### method 2

```python
tamil_marks = float(input("Enter your tamil mark: "))
english_marks = float(input("Enter your english mark: "))
maths_marks = float(input("Enter your maths mark: "))
science_marks = float(input("Enter your science mark: "))
social_marks = float(input("Enter your social mark: "))

mark = [tamil_marks,english_marks,maths_marks,science_marks,social_marks]
print(mark)
total_mark = sum(mark)
print(total_mark)

avg = total_mark/len(mark)  # 437/5  len - length is used to count of the element
print(avg)
```

# Odd or even

```python
num = 6                   # even

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
num = 21                # odd

if num % 2 == 1:
    print("Odd number")
else:
    print("Even number")
```

```python
num = 21                # odd

if num % 2 != 0:
    print("Odd number")
else:
    print("Even number")
```
