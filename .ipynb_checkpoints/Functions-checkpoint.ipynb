## Parameters and Aruguments

### Parameters is used to pass a input value in the function definition or function declaration -> def keyword

### Arguments is used to pass a value for a input in the function call or when calling a function 

```python
def sub():
    a = 5  # a -> 5
    b = 3   # b -> 3
    print(a-b)  # 5-3 -> 2
sub()
```

```python
def sub(a,b):   # -> parameters  # a -> 5 , b -> 3
    
    print(a-b)  # 5-3 -> 2
```

```python
sub(5,3)    # -> Arguments
```

```python
def maths(a,b,c,d):  # a -> 10, b-> 5, c-> 3, d -> 2

    print(a+b-c*d)   # 3*2 -> 6 # 10+5 -> 15 # 15-6 -> 9       # bodmus rule ( / * + - 

maths(10,5,3,2)
```

```python
def wild_animals(a):
    animal = input("Enter a king of animal: ")
    print(f"This {animal} has {a} legs")

def domestic_animals(a):
    animal = input("Enter a pet of animal: ")
    print(f"This {animal} has {a} legs")

def sea_animals(b):
    animal = input("Enter a king of sea animal: ")
    print(f"This {animal} has {b} legs")


a = int(input("Enter a number of legs for animals: "))
b = int(input("Enter a number of legs for sea_animals: "))

wild_animals(a)
domestic_animals(a)
sea_animals(b)
```

### Types of arguments

### 1.Positional arguments
### 2. Keyword arguments
### 3. Default arguments

### 1.Positional arugments

```python
def animals(name,leg):    # name -> lion , leg -> 4
    print(f"The {name} has {leg} legs")

animals(4,"lion")   # -> position arguments
```

### 2.Keyword Arguments

```python
def animals(name,leg):    # name -> lion , leg -> 4
    print(f"The {name} has {leg} legs")

animals(leg=4,name="lion")   # -> keyword arguments
```

### 3. Default arguments

```python
def animals(name,hair="medium",leg = 5):  # default arguments compulsory pass in last of the parameters
    print(f"The {name} has {leg} legs and have {hair} hair")

animals("lion")
```

```python
def animals(name,hair="medium",leg = 5):  # default arguments compulsory pass in last of the parameters
    print(f"The {name} has {leg} legs and have {hair} hair")

animals(name="lion",leg=4,hair="long")
```
