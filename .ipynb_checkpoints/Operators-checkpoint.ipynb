## Logical operator

### And,Or,Not

```python
statement1 = 8
statement2 = 5
print(statement1,statement2)
```

```python
print(statement1==statement2)
```

```python
print(statement1==8 and statement2==5)  # 8==7 f and 5==5 T
```

```python
print(statement1==8 or statement2==4)  # T or F
```

```python
print(statement1!=8)
```

```python
print(not(True))
```

```python
print(not(statement1==8))
```

```python
age = 16
voter_id = False

if age >= 18 or voter_id:
    print("You can vote!")
else:
    print("You cannot vote")
```

## Assigment operator or Agumented assignment operator 
### = or Agumented assignment operator means combination of both arithmetic operator and assigment opertor

```python
age = 20
print(age)
```

```python
age = age + 2  # 20 + 2 = 22
print(age)
```

```python
age += 2    # 20 + 2 = 22
print(age)
```

```python
age -= 3
print(age)
```

## Membership operator
### in and not in

```python
veg = ["carrot","ladies finger","potato","tomato","beetroot"]
print(veg)
```

```python
print("tomato" in veg)
```

```python
print("carrot" not in veg)
```

## Identity operator
### is and is not

```python
name = 'vijay'
name1 = 'vijay'
Name = 'ajith'
age = 50
print(name,name1,Name,age)
```

```python
print(id(name))
print(id(name1))
print(id(Name))
print(id(age))
```

```python
print(name is age)
```

```python
print(name is not age)
```

## Ternary or Conditional operator

### syntax value_if_true if condition value_if_false

```python
age = 16
voter_id = False

if age >= 18 or voter_id:
    print("You can vote!")
else:
    print("You cannot vote")
```

```python
age = 16
voter_id = False

vote = ["you can vote!" if age>=18 or voter_id else "You cannot vote"]
print(vote)
```
