# Python DataTypes

## Kind of Data

## simple data types - cat 

### integer - int used for whole numbers   eg : 1,2,3
### string - str used for alphabets character    eg: "" or 'Python'
### float - float used for decimal point numbers   eg: 1.0, 23.00
### boolean - bool used for checking like ouput as true or false   eg: true - 1 and false - 0

## complex data types or data structure

### list - lst used for collection of items within []  eg: [1,"python","true",12.00]
### tuple - tup used for collection of items within ()  eg: (1,"python","true",12.00) 
### set - set used for collection of items but in unorder within {}   eg:{1,"python","true",12.00}
### dictionary - dict used for collection of items with key and value pairs with {}   eg:{"s_no":1,"language":"python"}
