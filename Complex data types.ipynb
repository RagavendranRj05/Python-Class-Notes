## set operations

### Union
### Intesection
### Difference
### Symetric difference

```python
SetA = {1,2,3,4,5,'lion'}
SetB = {6,7,8,9,0}
SetC = {1,3,5,4,6,8,'lion'}
SetD = {1,2,3,'lion','tiger','girafee','elephent','buffalo','camel'}
print(SetA)
print(SetB)
print(SetC)
print(SetD)
```

### union or update is used to join the one or more than sets and operator |

```python
SetA.update(SetB,SetD)
print(SetA)
```

```python
a=SetA.union(SetB,SetD)
print(a)
```

```python
a = SetA | SetB | SetD   # | union operator symbol
print(a)
```

### Intersection is used to return the common values of both sets or more than sets  and operator &

```python
print(SetA)
```

```python
print(SetC)
```

```python
print(SetD)
```

```python
b = SetA.intersection(SetC,SetD) 
print(b)
```

```python
Inter = SetA & SetC & SetD
print(Inter)
```

### Difference means return a element in First set A but not in another SetB and operator -

```python
print(SetA)
```

```python
print(SetC)
```

```python
dif=SetC.difference(SetA)
print(dif)
```

```python
Inter1 = SetA & SetC
print(Inter1)
```

```python
Interop = SetC - SetA
print(Interop)
```

### Symmetric Difference means return a element in either A or B but not present in both Set and operator ^

```python
print(SetA)
```

```python
print(SetB)
```

```python
print(SetC)
```

```python
print(SetD)
```

```python
Sym = SetC.symmetric_difference(SetD)
print(Sym)
```

```python
sym1 = SetA.symmetric_difference(SetC)
print(sym1)
```

```python
syms = SetA ^ SetC
print(syms)
```
